```python
# Fill in these values
sampling_period = 2 # in ms
num_timesteps = 150

average1 = sum(data['neuron1'],2)/100
average2 = sum(data['neuron2'],2)/100
average3 = sum(data['neuron3'],2)/100
average4 = sum(data['neuron4'],2)/100

velocity_direction = data['stim']

plt.figure(figsize=(9.6,7.2))
plt.subplot(231)
plt.plot(velocity_direction, average1)
plt.xlabel('velocity direction in degrees')
plt.ylabel('Firing Rate in Hz.')
plt.title('neuron 1')
plt.subplot(232)
plt.plot(velocity_direction, average2)
plt.xlabel('velocity direction in degrees')
plt.ylabel('Firing Rate in Hz.')
plt.title('neuron 2')
plt.subplot(233)
plt.plot(velocity_direction, average3)
plt.xlabel('velocity direction in degrees')
plt.ylabel('Firing Rate in Hz.')
plt.title('neuron 3')
plt.subplot(234)
plt.plot(velocity_direction, average4)
plt.xlabel('velocity direction in degrees')
plt.ylabel('Firing Rate in Hz.')
plt.title('neuron 4')
plt.subplot(235)
plt.plot(velocity_direction, average1, velocity_direction, average2, velocity_direction, average3, velocity_direction, average4)
plt.xlabel('velocity direction in degrees')
plt.ylabel('Firing Rate in Hz.')
plt.title('all')

plt.subplots_adjust(top=0.88, bottom=0.08, left=0.10, right=0.95, hspace=0.45,
                    wspace=0.45)

plt.suptitle('The mean firing rate of the neuron as a function of the stimulus.')
plt.show()
```
